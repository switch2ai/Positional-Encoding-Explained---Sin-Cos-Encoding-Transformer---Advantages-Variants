####Positional Encoding :

```python
-- In language order is important.

Previous architectures like RNN, LSTM :
  -- Order is maintaned by passing input time step wise
  -- Because of this, we can not process input parallely
```

```python
## Positional Encoding :
-- We create unique vectors for each position
-- Add this vector to embedding of token at that position

## Properties of positional Encoding :
-- Vector should be unique for each position
-- Dimension of vector must be equal to dimension of embedding vector
-- Value of positional encoding vector should be small, so it won't change the meaning of input vector
```

```python
## Methods to create positional encoding :
1. Learning positional encoding during training :
      -- Training time will be more
      -- If we get sequence longer than, training length we can not process it

2. Sinusoidal encoding : Sin & Cos
```
